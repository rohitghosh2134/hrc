# Aim:
 To find the clear date of the invoice entered using data analysis and machine learning of previous entries

## Data Dictionary
    business_code:  company code of the account
    cust_number: customer number given to all the customers of the account
    name_customer: name of the customer
    clear_date: date on which the customer clears an invoice, or in simple terms, makes the full payment
    business_year: the year in which the invoice was created
    doc_id: a unique identifier of an invoice
    posting_date: the date on which the particular invoice was entered in the ERP database
    document_create_date: the date on which the invoice document was created
    document_create_date.1: normalized version of document_create_date (we'll use this to split the
    data)
    due_in_date: the date on which the customer is expected to clear an invoice
    invoice_currency: the currency of the invoice amount in the document of the invoice
    document type: represents the type of document, eg, D1 represents invoice
    posting_id: key indicator to identify whether on AR terms is invoice, deduction, credit memo based
    on its value. Applicable for SAP ERP
    area_business: business area in sap is defined as an organisational area within the financial
    accounting module
    total_open_amount: the amount that is yet to be paid for that invoice
    baseline_create_date: the date on which the invoice was created
    cust_payment_terms: business terms and agreements between customers and accounts on
    discounts and days of payments
    invoice_id: unique number assigned when a seller creates an invoice
    isOpen: indicator of whether an invoice is open or closed. isOpen = 1, means that the invoice is still
    open

### Assumption
Since we need to find out clear date ,we need to have delay column which will depend upon the rest of the columns. If we correctly predict the delay between the due date and clear date, we can ultimatley find out the clear date

Checking to find out any missing values, duplicate values and making a delay column first

```python
#With the help of datatype of each column, making a delay column
ddate=pd.to_datetime(df['due_in_date'],format='%Y%m%d')
cdate=pd.to_datetime(df['clear_date'])
df['delay']=(cdate-ddate).dt.days
```

# Encoding

```python
#One hot encoding
def binary_encoder(df,column):
    df=df.copy()
    df=pd.get_dummies(df,columns=[column])
    return df
```

```python
#label Encoding
def label(df,column):
    df=df.copy()
    lbe=LabelEncoder()
    df[column]=lbe.fit_transform(df[column])
    return df
```

# Preprocessing Bigdata dataset

```python
#checking any missing values in bigdata
bigdata.isna().sum()
```

```python
#checking for correlation between columns
sns.heatmap(bigdata.corr(),annot=True)
```

As per the heatmap, we see that there is no relation between posting id,isOpen and area business columns with others.
So we can drop them.
We can also drop document_create_date,baseline_create_date as the normalised version of it is already present.
Note that, business code column, cust _payment_terms,invoice_currency,document type which maybe helpful for delay prediction are missing, we need to label encode those columns.

```python
#dropping unwanted columns
bigdata=bigdata.drop(['posting_id','area_business','document_create_date','isOpen','baseline_create_date'],axis=1)
```

```python
bigdata.shape
```

```python
#checking na values in bigdata
bigdata.isna().sum()
```

```python
#dropping any na values should not exist here
bigdata=bigdata.dropna()
```

```python
#dropping clear_date column because it is no longer required
bigdata=bigdata.drop(['clear_date'],axis=1)
```

```python
#checking again
bigdata.isna().sum()
```

```python
#dropping duplicate items from doc_id, invoice_id which are unique as stated in the data dictionary
bigdata=bigdata.drop_duplicates(subset=['doc_id','invoice_id'],keep='first')
bigdata.shape
```

```python
#now checking if any unique values are left
{column: len(bigdata[column].unique()) for column in bigdata.columns}
```

```python
#dropping doc_id,invoice_id columns because everything is unique and it may not contribute to prediction
#also dropping name_customer column which is identified by its cust_number,posting date due to normalised form of document_create_date.1
bigdata=bigdata.drop(['doc_id','invoice_id','name_customer','posting_date'],axis=1)
```

```python
#now checking again if any unique values are left
{column: len(bigdata[column].unique()) for column in bigdata.columns}
```

```python
bigdata.dtypes
```

```python
#label endcoding all the business_code, cust_number,cust_payment_terms  columns
bigdata=label(bigdata,'cust_number')
bigdata=label(bigdata,'business_code')
bigdata=label(bigdata,'cust_payment_terms')
```

```python
#One Hot Encoding the invoice_currency,document type
bigdata=binary_encoder(bigdata,'invoice_currency')
bigdata=binary_encoder(bigdata,'document type')
```
